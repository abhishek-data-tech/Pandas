```python
# find the most number of matches plsyed by any teams.

number_matches_teams =ipl_matches["team1"].value_counts() + ipl_matches["team2"].value_counts()
number_matches_teams.sort_values(ascending=False).head(5)
```

```python
# drop_duplicates()

ipl_matches.drop_duplicates(subset="city").shape
```

```python
ipl_matches["date"] = pd.to_datetime(ipl_matches["date"])
```

```python
ipl_matches["season_year"] = ipl_matches["date"].dt.year
```

```python
ipl_matches.drop_duplicates(subset=["season_year"],keep="last")[["season_year","winner"]].sort_values(by="season_year")
```

```python
#Merge Joining dataframe

new_df = ipl_each_ball.merge(ipl_matches,left_on="id",right_on="id")
new_df.groupby(["season_year","batsman"])["batsman_runs"].sum().reset_index().sort_values("batsman_runs",ascending=False).drop_duplicates(subset=["season_year"],keep='first').sort_values("season_year")
```
